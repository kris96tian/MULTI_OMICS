```python
import matplotlib.pyplot as plt
import seaborn as sns
# Basic EDA
print(transposed_.describe().to_markdown(numalign="left", stralign="left"))
```

```python
transposed_df_control = transposed_
transposed_df_control
```

```python
import scipy.cluster.hierarchy as sch

# Calculate the correlation matrix
correlation_matrix = transposed_df_control.corr(method='spearman')  # You can use Pearson correlation as well


# Create a dendrogram for hierarchical clustering
plt.figure(figsize=(12, 6))
dendrogram = sch.dendrogram(sch.linkage(correlation_matrix, method='ward'))
plt.title("Dendrogram for Metabolite Clustering")
plt.xlabel("Metabolite")
plt.ylabel("Distance")
plt.show()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.clustermap(correlation_matrix, cmap='coolwarm', annot=True)  # clustermap combines heatmap with dendrogram
plt.title('Correlation Heatmap and Clustering (Control Samples)')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Perform PCA
pca = PCA()
pca_result = pca.fit_transform(transposed_df_control)

# Calculate explained variance for each PC
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained variance ratios: {explained_variance_ratio}")

# Plot the first two principal components
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot (Control Samples)')
plt.show()
```

```python
# Find metabolites with the largest variance across samples
variances = transposed_df_control.var(axis=1)
top_variable_metabolites = variances.nlargest(10)  # Get top 10 most variable metabolites
print("Top 10 Most Variable Metabolites:")
print(top_variable_metabolites.to_markdown(numalign="left", stralign="left"))
```

```python
from sklearn.covariance import EllipticEnvelope

# Fit an Elliptic Envelope outlier detection model
outlier_detector = EllipticEnvelope(contamination=0.05)  # You can adjust the contamination parameter
outlier_detector.fit(pca_result)  # Fit to PCA results

# Predict outliers
outliers = outlier_detector.predict(pca_result)
outlier_indices = np.where(outliers == -1)[0]  # Get indices of outliers

# Get the sample IDs of the outliers
outlier_samples = transposed_df_control.index[outlier_indices]

print("Potential Outlier Samples:")
for sample in outlier_samples:
    print(sample)
```

```python
import re

# Regular expression to match metabolites in the TG(X:Y) format
pattern = r"TG\((\d+):(\d+)\)"

# Get all metabolite names
metabolites = transposed_df_control.index

# Create the list of metabolite pairs
metabolite_pairs = []
for met1 in metabolites:
    if re.match(pattern, met1):
        for met2 in metabolites:
            if re.match(pattern, met2) and met1 != met2:  # Avoid ratios with the same metabolite
                if int(re.search(pattern, met1).group(1)) - int(re.search(pattern, met2).group(1)) == 2:
                  metabolite_pairs.append((met1, met2))


print(metabolite_pairs)
```
